```python
# Supervised Learing - Machine Learning - Tree Based Algorithms - Output of these
# algorithms is a Tree Model. Tree will have Root Node(Beginnine Node), Decision
# Node (Splitting Node) and Terminal Node or Leaf node(End Point)

# Decision Tree is a Single Tree Model which is a binary splitting tree and 
# recursive partitioning tree. Recurusive Partitioning is where tree splits on
# previous node number of observations.

# Once all observations classified into terminal node or leaf node, tree growth will
# stop. As the tree grows down number of observations will reduce

# Preditions are based on rules generated from tree. Tree must be read Top Down.

# Decision Tree is both Regression and classification Algorithm. Also used as
# Unsupervised Algorithm as it provides feature/variable importance as part of
# output.

# Tree Splitting methods - Classification Models
# 1) Gini Impurity Method - 1-sum(prob(i))^2
# Gini is calculated for all variables(X's) with dependent variable(y) and variable
# that has lowest gini is first split and ascending order till 0.40. Gini greater
# than 0.40, variable is deleted from tree as it is considered Insignificant Variable

# 2) Entropy & Information Gain Method - Entropy = sum-(prob(i) * Log(prob(i),2))
# Information Gain = Entropy of Dependent Variable - Entropy of Independent Variable
# Information Gain is calculated for all variables with dependent variable and
# variable that has highest information gain is first split and descending order
# till 0.50. Iformation gain less than 0.50, variable deleted from tree.

# Decision Tree is a non parametric algorithm that works with any type of data including
# data with skewness and outliers. Algorithm best suited for large scale data and
# requires less computing power

# Only disadvantage is Decision Tree overfits most of times (>0.95 - R Square or
# Accuracy) . Overfitting is controlled by reducing max_depth=
```

```python
#Split Data into dependent variable (y) and Independent Variable(X)
y=bigmarttrain.Item_Outlet_Sales
X=bigmarttrain.drop('Item_Outlet_Sales',axis=1)
```

```python
X=pd.get_dummies(X, columns=['Item_Fat_Content', 'Item_Type',
       'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type',
       'Outlet_Type', 'Item_Code'])
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree=DecisionTreeRegressor()
```

```python
treemodel=tree.fit(X,y)
```

```python
treemodel.score(X,y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(tree,X,y)
```

```python
# Variable Importance
pd.DataFrame(X.columns,treemodel.feature_importances_).sort_index(ascending=False)
```
